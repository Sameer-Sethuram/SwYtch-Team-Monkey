```python
import pandas as pd
import os
import glob
```

```python
# === Step 1: Load Dataset ===
DATA_PATH = "C:\git\SwYtch-Team-Monkey\Raw_Data\Indeed_Hiring_Lab\Sectors\seasonally_unadjusted"

# Get all CSV files in the directory
csv_files = glob.glob(os.path.join(DATA_PATH, "*.csv"))

print(f"Found {len(csv_files)} CSV files:")

for file in csv_files:
    print(f"  - {os.path.basename(file)}")
```

```python
# Now let's read all CSV files and combine them into one dataframe
dataframes = []

for file in csv_files:
    # Read each CSV file
    df_temp = pd.read_csv(file)
    
    # Add a column to identify which file the data came from
    df_temp['source_file'] = os.path.basename(file)
    
    # Add to our list of dataframes
    dataframes.append(df_temp)
    
    print(f"Loaded {file}: {df_temp.shape[0]} rows, {df_temp.shape[1]} columns")

# Combine all dataframes using pd.concat()
combined_df = pd.concat(dataframes, ignore_index=True)

print(f"\nCombined dataframe shape: {combined_df.shape}")
print(f"Total rows: {len(combined_df)}")
print(f"Total columns: {len(combined_df.columns)}")
```

```python
# Let's take a look at the combined dataframe
print("\nColumn names:")
print(combined_df.columns.tolist())

print("\nUnique source files:")
print(combined_df['source_file'].unique())

combined_df
```

```python
# === Step 2: Generate Data Dictionary ===
def infer_dtype(dtype):
    if pd.api.types.is_string_dtype(dtype):
        return "String"
    elif pd.api.types.is_numeric_dtype(dtype):
        return "Numeric"
    elif pd.api.types.is_bool_dtype(dtype):
        return "Boolean"
    elif pd.api.types.is_datetime64_any_dtype(dtype):
        return "Datetime"
    else:
        return "Other"

dictionary_lines = []
for col in combined_df.columns:
    dtype = infer_dtype(combined_df[col].dtype)
    example = combined_df[col].dropna().iloc[0] if not combined_df[col].dropna().empty else "N/A"
    dictionary_lines.append(f"    - *{col}* → {dtype} (e.g., {example})  # TODO: Add description")
```

```python
# === Step 3: Print README Section ===
print(dictionary_lines)
```
